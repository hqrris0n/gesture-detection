# Imports

```python
from ultralytics import YOLO
import cv2
import win32api
import win32con
import time
```

# Variable setup

```python
model = YOLO("myModels/mainv1/best.pt")
lastId = 0
frameCounter = 0
```

# Keyboard input method

```python
def virtualKey(key):
    win32api.keybd_event(key, 0, 0, 0)  # Key down
    time.sleep(0.1)  # Wait for a short time
    win32api.keybd_event(key, 0, win32con.KEYEVENTF_KEYUP, 0)  # Key up

def takeInput(id):
    global lastId
    global frameCounter
    frameDelay = 7

    # Adding a frame delay to ensure keypresses
    if not id == lastId:
        frameCounter = 0
        lastId = id
    elif frameCounter < frameDelay:
        frameCounter += 1
    else:
        frameCounter = 0

        if id == 0:
            virtualKey(0x4F)
        elif (id == 1):
            virtualKey(0x45)
        elif (id == 2):
            virtualKey(0x44)
        elif (id == 3):
            virtualKey(0x4C)
        elif (id == 4):
            virtualKey(0x41)
        elif (id == 5):
            virtualKey(0x4A)
        elif (id == 6):
            virtualKey(0x49)
        elif (id == 7):
            virtualKey(0x51)
        elif (id == 8):
            virtualKey(0x4B)
        elif (id == 9):
            virtualKey(0x53)
        elif (id == 10):
            virtualKey(0x50)
        elif (id == 11):
            virtualKey(0x57)
```

# Image detection

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 tracking on the frame, persisting tracks between frames
        results = model.predict(frame, conf=0.5, max_det=1)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()
        
        # Extract results
        for result in results:
            # Access the class IDs tensor and convert it to a NumPy array
            class_ids = result.boxes.cls.cpu().numpy()

            # Print the class IDs of detected objects
            for cls_id in class_ids:
                takeInput(cls_id)

        # Display the annotated frame
        cv2.imshow("Controller Camera", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        print("Failed to get video")
        break
```

```python
# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

# Camera Testing

```python
max_cameras = 2
available_cameras = []
for i in range(max_cameras):
    cap = cv2.VideoCapture(i)
    if cap.isOpened():
        print(f"Camera index {i} is available.")
        available_cameras.append(i)
        cap.release()
    else:
        print(f"Camera index {i} is not available.")
print("Available camera indices:", available_cameras)
```
